# 🎓 Institute Dashboard — Dataset Overview

**Datasets:** Admissions · Students · Attendance · Courses · Staff · Schedule · Classrooms · Placement  
**Purpose:** High-level visual overview of all institute datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# ── Color palette ──────────────────────────────────────────────
COLORS = ['#4C72B0','#DD8452','#55A868','#C44E52','#8172B2',
          '#937860','#DA8BC3','#8C8C8C','#CCB974','#64B5CD']
ACCENT  = '#4C72B0'
ORANGE  = '#DD8452'
GREEN   = '#55A868'
RED     = '#C44E52'

plt.rcParams.update({'figure.dpi': 120, 'font.family': 'DejaVu Sans',
                     'axes.spines.top': False, 'axes.spines.right': False})

# ── Load all datasets ──────────────────────────────────────────
BASE = 'https://raw.githubusercontent.com/'
admission   = pd.read_csv('admission.csv')
attendance  = pd.read_csv('attendance_current_month.csv')
classroom   = pd.read_csv('classroom.csv')
courses     = pd.read_csv('course_dataset.csv')
placement   = pd.read_csv('placement.csv')
staff       = pd.read_csv('staff_dataset.csv')
schedule    = pd.read_csv('staff_schedule_with_weekly.csv')
students    = pd.read_csv('students.csv')

print('✅ All datasets loaded')
for name, df in [('Admission',admission),('Attendance',attendance),
                 ('Classroom',classroom),('Courses',courses),
                 ('Placement',placement),('Staff',staff),
                 ('Schedule',schedule),('Students',students)]:
    print(f'  {name:<12} → {df.shape[0]:>4} rows × {df.shape[1]:>2} cols')
```

## 1 · Dataset Snapshot — Shape & Null Overview

```python
datasets = {
    'Admission': admission, 'Attendance': attendance, 'Classroom': classroom,
    'Courses': courses,     'Placement': placement,   'Staff': staff,
    'Schedule': schedule,   'Students': students
}

summary = pd.DataFrame([
    {'Dataset': k, 'Rows': v.shape[0], 'Columns': v.shape[1],
     'Nulls': v.isnull().sum().sum(),
     'Null %': round(v.isnull().sum().sum() / (v.shape[0]*v.shape[1]) * 100, 1)}
    for k, v in datasets.items()
])

fig, axes = plt.subplots(1, 3, figsize=(16, 4))
fig.suptitle('Dataset Snapshot', fontsize=14, fontweight='bold', y=1.02)

# Rows
axes[0].barh(summary['Dataset'], summary['Rows'], color=COLORS[:8])
axes[0].set_title('Rows per Dataset', fontweight='bold')
axes[0].set_xlabel('Row Count')
for i, v in enumerate(summary['Rows']):
    axes[0].text(v+2, i, str(v), va='center', fontsize=9)

# Columns
axes[1].barh(summary['Dataset'], summary['Columns'], color=COLORS[:8])
axes[1].set_title('Columns per Dataset', fontweight='bold')
axes[1].set_xlabel('Column Count')
for i, v in enumerate(summary['Columns']):
    axes[1].text(v+0.1, i, str(v), va='center', fontsize=9)

# Null %
bar_colors = [RED if x > 10 else ORANGE if x > 5 else GREEN for x in summary['Null %']]
axes[2].barh(summary['Dataset'], summary['Null %'], color=bar_colors)
axes[2].set_title('Null Data %', fontweight='bold')
axes[2].set_xlabel('Null %')
for i, v in enumerate(summary['Null %']):
    axes[2].text(v+0.3, i, f'{v}%', va='center', fontsize=9)
patches = [mpatches.Patch(color=GREEN,label='<5%'),
           mpatches.Patch(color=ORANGE,label='5-10%'),
           mpatches.Patch(color=RED,label='>10%')]
axes[2].legend(handles=patches, fontsize=8, loc='lower right')

plt.tight_layout()
plt.show()
display(summary.set_index('Dataset'))
```

## 2 · Admissions Overview

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Admissions Overview', fontsize=15, fontweight='bold')

# Status
status_counts = admission['status'].value_counts()
axes[0,0].pie(status_counts, labels=status_counts.index, autopct='%1.1f%%',
              colors=COLORS[:len(status_counts)], startangle=90)
axes[0,0].set_title('Admission Status')

# Course distribution
course_counts = admission['course'].value_counts().head(8)
axes[0,1].barh(course_counts.index, course_counts.values, color=COLORS[:8])
axes[0,1].set_title('Admissions by Course (Top 8)')
axes[0,1].set_xlabel('Count')

# Mode
mode_counts = admission['mode'].value_counts()
axes[0,2].bar(mode_counts.index, mode_counts.values, color=COLORS[:len(mode_counts)])
axes[0,2].set_title('Admission Mode')
axes[0,2].set_ylabel('Count')
for i, v in enumerate(mode_counts.values):
    axes[0,2].text(i, v+1, str(v), ha='center', fontweight='bold')

# Qualification
qual_counts = admission['qualification'].value_counts()
axes[1,0].bar(qual_counts.index, qual_counts.values, color=COLORS[:len(qual_counts)])
axes[1,0].set_title('Qualification Level')
axes[1,0].set_ylabel('Count')
axes[1,0].tick_params(axis='x', rotation=20)

# Location
loc_counts = admission['location'].value_counts().head(8)
axes[1,1].barh(loc_counts.index, loc_counts.values, color=COLORS[:8])
axes[1,1].set_title('Top Locations')
axes[1,1].set_xlabel('Count')

# Attended by (counsellor)
counsel_counts = admission['attended_by'].value_counts().head(8)
axes[1,2].bar(counsel_counts.index, counsel_counts.values, color=COLORS[:8])
axes[1,2].set_title('Admissions by Counsellor')
axes[1,2].set_ylabel('Count')
axes[1,2].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

## 3 · Student Demographics

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Student Demographics & Enrollment', fontsize=15, fontweight='bold')

# Age distribution
axes[0,0].hist(students['age'], bins=15, color=ACCENT, edgecolor='white')
axes[0,0].set_title('Age Distribution')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Students')
axes[0,0].axvline(students['age'].mean(), color=RED, linestyle='--',
                  label=f'Mean: {students["age"].mean():.1f}')
axes[0,0].legend()

# Education level
edu_counts = students['education_level'].value_counts()
axes[0,1].pie(edu_counts, labels=edu_counts.index, autopct='%1.1f%%',
              colors=COLORS[:len(edu_counts)], startangle=90)
axes[0,1].set_title('Education Level')

# Mode
mode_counts = students['mode'].value_counts()
axes[0,2].pie(mode_counts, labels=mode_counts.index, autopct='%1.1f%%',
              colors=[ACCENT, ORANGE], startangle=90)
axes[0,2].set_title('Online vs Offline')

# Course enrolment
course_enroll = students['course'].value_counts().head(8)
axes[1,0].barh(course_enroll.index, course_enroll.values, color=COLORS[:8])
axes[1,0].set_title('Enrolment by Course (Top 8)')
axes[1,0].set_xlabel('Students')

# Duration
duration_counts = students['duration'].value_counts()
axes[1,1].bar(duration_counts.index, duration_counts.values, color=COLORS[:len(duration_counts)])
axes[1,1].set_title('Course Duration')
axes[1,1].set_ylabel('Count')
axes[1,1].tick_params(axis='x', rotation=20)

# Internship
intern_counts = students['internship'].value_counts()
axes[1,2].bar(intern_counts.index, intern_counts.values,
              color=[GREEN if x != 'No' else RED for x in intern_counts.index])
axes[1,2].set_title('Internship Status')
axes[1,2].set_ylabel('Count')
axes[1,2].tick_params(axis='x', rotation=25)

plt.tight_layout()
plt.show()
```

## 4 · Attendance Analysis

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Attendance — Current Month Analysis', fontsize=15, fontweight='bold')

att = attendance.copy()

# Attendance % distribution
axes[0,0].hist(att['Attendance %'], bins=10, color=ACCENT, edgecolor='white')
axes[0,0].set_title('Attendance % Distribution')
axes[0,0].set_xlabel('Attendance %')
axes[0,0].set_ylabel('Students')
axes[0,0].axvline(75, color=RED, linestyle='--', linewidth=2, label='75% threshold')
axes[0,0].legend()

# Status
status_counts = att['Status'].value_counts()
colors_status = [GREEN if s=='Regular' else ORANGE if s=='Warning' else RED
                 for s in status_counts.index]
axes[0,1].bar(status_counts.index, status_counts.values, color=colors_status)
axes[0,1].set_title('Attendance Status')
axes[0,1].set_ylabel('Students')
for i, v in enumerate(status_counts.values):
    axes[0,1].text(i, v+0.5, str(v), ha='center', fontweight='bold')

# Attendance by course
course_att = att.groupby('course')['Attendance %'].mean().sort_values()
bar_colors = [RED if x<75 else ORANGE if x<85 else GREEN for x in course_att.values]
axes[0,2].barh(course_att.index, course_att.values, color=bar_colors)
axes[0,2].set_title('Avg Attendance % by Course')
axes[0,2].set_xlabel('Attendance %')
axes[0,2].axvline(75, color=RED, linestyle='--')

# Present vs Absent
avg_present = att['Present'].mean()
avg_absent  = att['Absent'].mean()
axes[1,0].bar(['Avg Present', 'Avg Absent'], [avg_present, avg_absent],
              color=[GREEN, RED])
axes[1,0].set_title('Avg Present vs Absent Days')
axes[1,0].set_ylabel('Days')
for i, v in enumerate([avg_present, avg_absent]):
    axes[1,0].text(i, v+0.05, f'{v:.1f}', ha='center', fontweight='bold')

# Scatter: Present days vs Attendance %
scatter_colors = [GREEN if s=='Regular' else ORANGE if s=='Warning' else RED
                  for s in att['Status']]
axes[1,1].scatter(att['Present'], att['Attendance %'],
                  c=scatter_colors, alpha=0.6, edgecolors='none', s=40)
axes[1,1].set_title('Present Days vs Attendance %')
axes[1,1].set_xlabel('Days Present')
axes[1,1].set_ylabel('Attendance %')
axes[1,1].axhline(75, color=RED, linestyle='--', label='75% line')
axes[1,1].legend(fontsize=8)

# Top absentees (bottom 10)
low_att = att.nsmallest(10, 'Attendance %')[['student_name','Attendance %']]
axes[1,2].barh(low_att['student_name'], low_att['Attendance %'], color=RED)
axes[1,2].set_title('Bottom 10 — Lowest Attendance')
axes[1,2].set_xlabel('Attendance %')
axes[1,2].axvline(75, color='black', linestyle='--')

plt.tight_layout()
plt.show()
```

## 5 · Courses & Staff

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Courses & Staff Overview', fontsize=15, fontweight='bold')

# Batches per course
axes[0,0].barh(courses['course'], courses['total_batches'], color=COLORS[:len(courses)])
axes[0,0].set_title('Batches per Course')
axes[0,0].set_xlabel('Total Batches')

# Avg hours per course
axes[0,1].barh(courses['course'], courses['avg_hours'], color=ORANGE)
axes[0,1].set_title('Avg Daily Hours per Course')
axes[0,1].set_xlabel('Avg Hours')

# Staff per course
axes[0,2].barh(courses['course'], courses['total_staff'], color=GREEN)
axes[0,2].set_title('Staff per Course')
axes[0,2].set_xlabel('Staff Count')

# Staff roles
role_counts = staff['role'].value_counts()
axes[1,0].bar(role_counts.index, role_counts.values, color=COLORS[:len(role_counts)])
axes[1,0].set_title('Staff by Role')
axes[1,0].set_ylabel('Count')
axes[1,0].tick_params(axis='x', rotation=30)

# Batches handled per staff
staff_sorted = staff.sort_values('batches_handling', ascending=True)
axes[1,1].barh(staff_sorted['name'], staff_sorted['batches_handling'], color=ACCENT)
axes[1,1].set_title('Batches per Staff Member')
axes[1,1].set_xlabel('Batches')

# Hours per day distribution
hours_counts = staff['hours_per_day'].value_counts().sort_index()
axes[1,2].bar(hours_counts.index.astype(str), hours_counts.values, color=COLORS[:len(hours_counts)])
axes[1,2].set_title('Hours per Day Distribution (Staff)')
axes[1,2].set_xlabel('Hours/Day')
axes[1,2].set_ylabel('Staff Count')

plt.tight_layout()
plt.show()
```

## 6 · Staff Weekly Schedule

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Staff Schedule Overview', fontsize=15, fontweight='bold')

# Weekly hours
sch = schedule.sort_values('weekly_hours')
axes[0].barh(sch['name'], sch['weekly_hours'], color=ACCENT)
axes[0].set_title('Weekly Hours per Staff')
axes[0].set_xlabel('Hours/Week')
axes[0].axvline(sch['weekly_hours'].mean(), color=RED, linestyle='--',
                label=f'Avg: {sch["weekly_hours"].mean():.0f}h')
axes[0].legend()

# Daily hours
axes[1].barh(sch['name'], sch['daily_hours'], color=ORANGE)
axes[1].set_title('Daily Hours per Staff')
axes[1].set_xlabel('Hours/Day')

# Slots heatmap
slot_cols = ['10:00-12:00','12:00-02:00','02:30-04:30','04:30-05:30 (Online)']
slot_usage = pd.DataFrame({
    col: [(x.strip() not in ['Free Period','free',''] if pd.notna(x) else False)
          for x in schedule[col]]
    for col in slot_cols
}, index=schedule['name'])

im = axes[2].imshow(slot_usage.values.astype(int), cmap='RdYlGn', aspect='auto',
                    vmin=0, vmax=1)
axes[2].set_xticks(range(len(slot_cols)))
axes[2].set_xticklabels(['10-12','12-2','2:30-4:30','4:30-5:30\n(Online)'], fontsize=8)
axes[2].set_yticks(range(len(schedule)))
axes[2].set_yticklabels(schedule['name'], fontsize=8)
axes[2].set_title('Slot Utilisation (Green=Assigned)')
plt.colorbar(im, ax=axes[2], shrink=0.6)

plt.tight_layout()
plt.show()
```

## 7 · Classroom Utilisation

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Classroom Utilisation', fontsize=15, fontweight='bold')

# Slot occupancy heatmap
slot_cols = ['10:00-12:00','12:00-02:00','02:30-04:30','04:30-05:30']
room_matrix = classroom.set_index('CID')[slot_cols]
# Encode: free=0, assigned=1
room_encoded = room_matrix.apply(lambda col: col.map(lambda x: 0 if pd.isna(x) or x.strip()=='' else 1))

im = axes[0].imshow(room_encoded.values, cmap='YlGn', aspect='auto', vmin=0, vmax=1)
axes[0].set_xticks(range(len(slot_cols)))
axes[0].set_xticklabels(slot_cols, rotation=20, fontsize=9)
axes[0].set_yticks(range(len(classroom)))
axes[0].set_yticklabels(classroom['CID'])
axes[0].set_title('Room × Slot Assignment')
# Annotate with batch names
for i, row in room_matrix.iterrows():
    for j, val in enumerate(row):
        idx = list(room_matrix.index).index(i)
        if pd.notna(val) and val.strip():
            axes[0].text(j, idx, val.strip(), ha='center', va='center', fontsize=7, color='navy')

# Occupancy capacity
axes[1].bar(classroom['CID'], classroom['Occupancy'], color=ACCENT)
axes[1].set_title('Room Occupancy Capacity')
axes[1].set_xlabel('Classroom')
axes[1].set_ylabel('Capacity')
for i, v in enumerate(classroom['Occupancy']):
    axes[1].text(i, v+0.2, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## 8 · Placement Analysis

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Placement Analysis', fontsize=15, fontweight='bold')

pl = placement.copy()
placed = pl[pl['Status'] == 'Placed']

# Placement status
status_counts = pl['Status'].value_counts()
pcolors = [GREEN if s=='Placed' else ORANGE if s=='In Progress' else RED
           for s in status_counts.index]
axes[0,0].bar(status_counts.index, status_counts.values, color=pcolors)
axes[0,0].set_title('Placement Status')
axes[0,0].set_ylabel('Students')
for i, v in enumerate(status_counts.values):
    axes[0,0].text(i, v+0.5, str(v), ha='center', fontweight='bold')

# Placement rate by course
course_total   = pl.groupby('Course').size()
course_placed  = pl[pl['Status']=='Placed'].groupby('Course').size()
course_rate    = (course_placed / course_total * 100).dropna().sort_values()
bar_c = [GREEN if x>=70 else ORANGE if x>=40 else RED for x in course_rate.values]
axes[0,1].barh(course_rate.index, course_rate.values, color=bar_c)
axes[0,1].set_title('Placement Rate by Course (%)')
axes[0,1].set_xlabel('Placement %')

# Package distribution
if not placed.empty:
    axes[0,2].hist(placed['Package (Rs.L)'].dropna(), bins=10,
                   color=GREEN, edgecolor='white')
    axes[0,2].axvline(placed['Package (Rs.L)'].mean(), color=RED, linestyle='--',
                      label=f'Mean: {placed["Package (Rs.L)"].mean():.1f}L')
    axes[0,2].set_title('Package Distribution (Placed Students)')
    axes[0,2].set_xlabel('Package (Rs. Lakh)')
    axes[0,2].set_ylabel('Count')
    axes[0,2].legend()

# Top hiring companies
company_counts = placed['Company'].value_counts().head(8) if not placed.empty \
                 else pl['Company'].value_counts().head(8)
axes[1,0].barh(company_counts.index, company_counts.values, color=COLORS[:8])
axes[1,0].set_title('Top Hiring Companies')
axes[1,0].set_xlabel('Placements')

# Top roles
role_counts = placed['Role'].value_counts().head(8) if not placed.empty \
              else pl['Role'].value_counts().head(8)
axes[1,1].barh(role_counts.index, role_counts.values, color=COLORS[:8])
axes[1,1].set_title('Top Job Roles')
axes[1,1].set_xlabel('Count')

# Placement officer performance
officer_counts = pl[pl['Status']=='Placed']['Placement Officer'].value_counts() if not placed.empty \
                 else pl['Placement Officer'].value_counts()
if not officer_counts.empty:
    axes[1,2].bar(officer_counts.index, officer_counts.values, color=ACCENT)
    axes[1,2].set_title('Placements by Officer')
    axes[1,2].set_ylabel('Count')
    axes[1,2].tick_params(axis='x', rotation=25)

plt.tight_layout()
plt.show()
```

## 9 · Executive KPI Summary

```python
# ── Compute KPIs ───────────────────────────────────────────────
total_students     = len(students)
total_admissions   = len(admission)
active_admissions  = (admission['status']=='Admitted').sum()
total_staff        = len(staff)
total_courses      = len(courses)
total_batches      = courses['total_batches'].sum()
avg_attendance     = attendance['Attendance %'].mean()
low_attendance     = (attendance['Attendance %'] < 75).sum()
placement_rate     = (placement['Status']=='Placed').sum() / len(placement) * 100
avg_package        = placement[placement['Status']=='Placed']['Package (Rs.L)'].mean()
online_students    = (students['mode']=='Online').sum()
offline_students   = (students['mode']=='Offline').sum()

# ── Draw KPI cards ─────────────────────────────────────────────
fig = plt.figure(figsize=(18, 6))
fig.patch.set_facecolor('#F0F4F8')
fig.suptitle('📊 Executive KPI Summary', fontsize=16, fontweight='bold', y=1.02)

kpis = [
    ('Total Students',    f'{total_students}',           ACCENT),
    ('Active Admissions', f'{active_admissions}',        GREEN),
    ('Total Staff',       f'{total_staff}',              ORANGE),
    ('Courses Offered',   f'{total_courses}',            COLORS[4]),
    ('Running Batches',   f'{total_batches}',            COLORS[5]),
    ('Avg Attendance',    f'{avg_attendance:.1f}%',      GREEN if avg_attendance>=75 else RED),
    ('< 75% Attendance',  f'{low_attendance} students',  RED),
    ('Placement Rate',    f'{placement_rate:.1f}%',      GREEN if placement_rate>=50 else ORANGE),
    ('Avg Package',       f'₹{avg_package:.1f}L' if not np.isnan(avg_package) else 'N/A', GREEN),
    ('Online Students',   f'{online_students}',          ACCENT),
    ('Offline Students',  f'{offline_students}',         ORANGE),
    ('Total Admissions',  f'{total_admissions}',         COLORS[6]),
]

cols = 6
rows = 2
for idx, (label, value, color) in enumerate(kpis):
    ax = fig.add_subplot(rows, cols, idx+1)
    ax.set_facecolor(color)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    ax.text(0.5, 0.62, value, ha='center', va='center', fontsize=20,
            fontweight='bold', color='white', transform=ax.transAxes)
    ax.text(0.5, 0.25, label, ha='center', va='center', fontsize=9,
            color='white', transform=ax.transAxes, wrap=True)
    for spine in ax.spines.values():
        spine.set_visible(False)

plt.tight_layout()
plt.show()
print('\n✅ Overview notebook complete!')
```
