```python
import streamlit as st
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
st.set_page_config(page_title="Student & Admission Dashboard", layout="wide")
```

```python
# ========================
# LOAD DATA
# ========================
@st.cache_data
def load_data():
    students = pd.read_csv(r"C:\Users\salmanul faris\Downloads\\students.csv")
    admissions = pd.read_csv(r"C:\Users\salmanul faris\Downloads\\admission.csv")
    return students, admissions

students, admissions = load_data()
```

```python
# Convert dates
students["enrollment_date"] = pd.to_datetime(students["enrollment_date"], errors='coerce')
admissions["admission_date"] = pd.to_datetime(admissions["admission_date"], errors='coerce')
```

```python
# ========================
# SIDEBAR FILTERS
# ========================
st.sidebar.header("Filters")

course_filter = st.sidebar.multiselect(
    "Select Course", students["course"].unique()
)

batch_filter = st.sidebar.multiselect(
    "Select Batch", students["batch_id"].unique()
)

mode_filter = st.sidebar.multiselect(
    "Select Mode", students["mode"].unique()
)
```

```python
# Apply filters
if course_filter:
    students = students[students["course"].isin(course_filter)]
    admissions = admissions[admissions["course"].isin(course_filter)]

if batch_filter:
    students = students[students["batch_id"].isin(batch_filter)]

if mode_filter:
    students = students[students["mode"].isin(mode_filter)]
    admissions = admissions[admissions["mode"].isin(mode_filter)]
```

```python
# ========================
# TITLE
# ========================
st.title("🎓 Students & Admissions Dashboard")
```

```python
# 1. STUDENT OVERVIEW
# ========================
st.header("📊 Student Overview")

col1, col2, col3 = st.columns(3)

col1.metric("Total Students", len(students))
col2.metric("Courses", students["course"].nunique())
col3.metric("Batches", students["batch_id"].nunique())
```

```python
# Age distribution
fig_age = px.histogram(students, x="age", nbins=20, title="Age Distribution")
st.plotly_chart(fig_age, use_container_width=True)
```

```python
# Mode distribution
fig_mode = px.pie(students, names="mode", title="Mode (Online/Offline)")
st.plotly_chart(fig_mode, use_container_width=True)
```

```python
course_data = students["course"].value_counts().reset_index()
course_data.columns = ["course", "count"]

fig_course = px.bar(
    course_data,
    x="course",
    y="count",
    labels={"course": "Course", "count": "Number of Students"},
    title="Course Distribution"
)

st.plotly_chart(fig_course, use_container_width=True)
```

```python
# 2. ENROLLMENT TREND
# ========================
st.header("📈 Enrollment Trend")

trend = admissions.copy()
trend["month"] = trend["admission_date"].dt.to_period("M").astype(str)

trend_data = trend.groupby("month").size().reset_index(name="Admissions")

fig_trend = px.line(
    trend_data,
    x="month",
    y="Admissions",
    markers=True,
    title="Monthly Admissions Trend"
)

st.plotly_chart(fig_trend, use_container_width=True)
```

```python
# 3. ADMISSION FUNNEL
# ========================
st.header("🔻 Admission Funnel")

funnel = admissions["status"].value_counts().reset_index()
funnel.columns = ["Stage", "Count"]

fig_funnel = px.funnel(funnel, x="Count", y="Stage", title="Admission Funnel")
st.plotly_chart(fig_funnel, use_container_width=True)
```

```python
# 4. COUNSELOR PERFORMANCE
# ========================
st.header("👨‍💼 Counselor Performance")

counselor_perf = admissions.groupby(["attended_by", "status"]).size().reset_index(name="count")

fig_counselor = px.bar(
    counselor_perf,
    x="attended_by",
    y="count",
    color="status",
    barmode="group",
    title="Counselor Performance"
)

st.plotly_chart(fig_counselor, use_container_width=True)

st.dataframe(counselor_perf)
```

```python
# 5. SOURCE & LOCATION
# ========================
st.header("🌍 Location Breakdown")

location_data = admissions["location"].value_counts().reset_index()
location_data.columns = ["Location", "Count"]

fig_location = px.bar(
    location_data,
    x="Location",
    y="Count",
    title="Student Location Distribution"
)

st.plotly_chart(fig_location, use_container_width=True)
```

```python
# EXTRA: EDUCATION LEVEL
# ========================
st.header("🎓 Education Level Breakdown")

edu_data = students["education_level"].value_counts().reset_index()
edu_data.columns = ["Education", "Count"]

fig_edu = px.pie(
    edu_data,
    names="Education",
    values="Count",
    title="Education Level"
)

st.plotly_chart(fig_edu, use_container_width=True)
```
