## Setup — Load Datasets & Helpers

```python

import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# ── Load all datasets (place CSVs in the same folder) ───────────────────────
students   = pd.read_csv('students.csv')
admission  = pd.read_csv('admission.csv')
attendance = pd.read_csv('attendance_current_month.csv')
placement  = pd.read_csv('placement.csv')
staff      = pd.read_csv('staff_dataset.csv')
schedule   = pd.read_csv('staff_schedule_with_weekly.csv')
courses    = pd.read_csv('course_dataset.csv')
classroom  = pd.read_csv('classroom.csv')


# ── Palette ──────────────────────────────────────────────────────────────────
PALETTE = ['#6366F1','#06B6D4','#F59E0B','#10B981','#EC4899','#EF4444','#8B5CF6','#14B8A6','#F97316']

def plotly_defaults(fig, h=340):
    fig.update_layout(
        height=h,
        plot_bgcolor='rgba(0,0,0,0)',
        paper_bgcolor='white',
        margin=dict(t=40, b=20, l=10, r=10),
        font=dict(family='Arial, sans-serif', size=11, color='#374151'),
        legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='left', x=0),
    )
    fig.update_xaxes(showgrid=False, showline=False)
    fig.update_yaxes(showgrid=True, gridcolor='rgba(99,102,241,.08)', showline=False)
    return fig

print('✅ All datasets loaded successfully')
for name, df in [('Students',students),('Admission',admission),('Attendance',attendance),
                 ('Placement',placement),('Staff',staff),('Schedule',schedule),
                 ('Courses',courses),('Classroom',classroom)]:
    print(f'  {name:<12} {df.shape[0]:>4} rows × {df.shape[1]:>2} cols')
```

## 1 · KPI Computation & Summary Table

```python

# ═══════════════════════════════════════════════════════════════════════════════
# KPI COMPUTATION  — pre-compute all headline numbers
# ═══════════════════════════════════════════════════════════════════════════════

placed_count   = (placement['Status'] == 'Placed').sum()
inprog_count   = (placement['Status'] == 'In Progress').sum()
admitted_count = (admission['status'] == 'Admitted').sum()
plc_rate       = round(placed_count / max(len(placement), 1) * 100)
adm_rate       = round(admitted_count / max(len(admission), 1) * 100)
avg_att        = attendance['Attendance %'].mean()
low_att_count  = (attendance['Attendance %'] < 75).sum()
avg_pkg        = placement[placement['Status']=='Placed']['Package (Rs.L)'].mean()
teach_staff    = (staff['department'] == 'Teaching').sum()
total_batches  = courses['total_batches'].sum()

slot_cols_ov = ['10:00-12:00','12:00-02:00','02:30-04:30','04:30-05:30']
total_slots    = len(classroom) * len(slot_cols_ov)
occupied_slots = classroom[slot_cols_ov].apply(
    lambda col: col.astype(str).str.strip().replace('', np.nan).notna().sum()
).sum()

kpis = [
    ('👩‍🎓', 'Total Students',   str(len(students)),                         '#6366F1'),
    ('📥',   'Admission Rate',   f'{adm_rate}%',                             '#06B6D4'),
    ('🎯',   'Placement Rate',   f'{plc_rate}%',                             '#10B981'),
    ('📊',   'Avg Attendance',   f'{avg_att:.1f}%',                          '#F59E0B'),
    ('⚠️',  'Below 75% Att.',   str(int(low_att_count)),                    '#EF4444'),
    ('💰',   'Avg Package',      f'Rs.{avg_pkg:.1f}L',                       '#8B5CF6'),
    ('🏫',   'Classroom Fill',   f'{int(occupied_slots/total_slots*100)}%',  '#14B8A6'),
    ('📚',   'Active Batches',   str(int(total_batches)),                    '#F97316'),
]

# ── Display KPI summary table ────────────────────────────────────────────────
kpi_df = pd.DataFrame(kpis, columns=['Icon','Metric','Value','Color'])
print('═'*50)
print('  INSTITUTE COMMAND CENTRE — KPI SNAPSHOT')
print('═'*50)
for _, row in kpi_df.iterrows():
    print(f"  {row['Icon']}  {row['Metric']:<22}  {row['Value']}")
print('═'*50)
```

## 2 · KPI Visual — Indicator Grid

```python

# ── KPI Visual — Indicator grid ─────────────────────────────────────────────
fig_kpi = make_subplots(
    rows=2, cols=4,
    specs=[[{'type':'indicator'}]*4]*2,
    vertical_spacing=0.15,
    horizontal_spacing=0.05,
)
kpi_colors = ['#6366F1','#06B6D4','#10B981','#F59E0B','#EF4444','#8B5CF6','#14B8A6','#F97316']

kpi_vals_num = [
    len(students), adm_rate, plc_rate, float(f'{avg_att:.1f}'),
    int(low_att_count), float(f'{avg_pkg:.1f}'), int(occupied_slots/total_slots*100), int(total_batches)
]
kpi_labels = ['Total Students','Admission Rate %','Placement Rate %','Avg Attendance %',
              'Below 75% Att.','Avg Package (L)','Classroom Fill %','Active Batches']

for idx in range(8):
    r, c = divmod(idx, 4)
    fig_kpi.add_trace(go.Indicator(
        mode='number',
        value=kpi_vals_num[idx],
        title=dict(text=kpi_labels[idx], font=dict(size=12, color='#6B7280')),
        number=dict(font=dict(size=32, color=kpi_colors[idx], family='Arial Black')),
    ), row=r+1, col=c+1)

fig_kpi.update_layout(
    height=220, paper_bgcolor='white',
    margin=dict(t=30, b=10, l=10, r=10),
    title=dict(text='🎓 Institute Command Centre — KPI Overview',
               font=dict(size=16, color='#1E1B4B'), x=0.5),
)
fig_kpi.show()
```

## 3 · Placement Pipeline Funnel

```python

# ═══════════════════════════════════════════════════════════════════════════════
# ROW 2B — Placement Pipeline Funnel
# ═══════════════════════════════════════════════════════════════════════════════

pipeline_stages = [
    ('Total Students',        len(students)),
    ('Appeared for Placement', len(placement)),
    ('In Progress',           int(inprog_count)),
    ('Successfully Placed',   int(placed_count)),
]

fig_funnel = go.Figure(go.Funnel(
    y=[s for s, _ in pipeline_stages],
    x=[v for _, v in pipeline_stages],
    textinfo='value+percent initial',
    marker=dict(
        color=['#6366F1','#06B6D4','#F59E0B','#10B981'],
        line=dict(color='white', width=1.5),
    ),
    connector=dict(line=dict(color='#E0E7FF', width=1.5)),
    opacity=0.92,
))

fig_funnel.update_layout(
    title=dict(text='🔽 Placement Pipeline — Students to Placed',
               font=dict(size=14, color='#1E1B4B'), x=0),
    margin=dict(t=50, b=20, l=10, r=10),
    paper_bgcolor='white', plot_bgcolor='rgba(0,0,0,0)',
)
plotly_defaults(fig_funnel, 380)
fig_funnel.show()
```

## 4 · Row 3 — Three cross-dataset panels
- **Counsellor Conversion Rate** — enquiries vs admitted per counsellor
- **Attendance Risk Bands** — students bucketed Critical / Warning / Regular / Excellent
- **Staff Workload vs Capacity** — batch load % against 3-batch max

```python

# ═══════════════════════════════════════════════════════════════════════════════
# ROW 3 — Counsellor Conversion · Attendance Risk · Staff Workload
# ═══════════════════════════════════════════════════════════════════════════════

fig_row3 = make_subplots(
    rows=1, cols=3,
    subplot_titles=[
        '🎯 Counsellor Conversion Rate',
        '⚠️ Attendance Risk Distribution',
        '👩‍🏫 Staff Workload vs Capacity',
    ],
    column_widths=[0.36, 0.32, 0.32],
    horizontal_spacing=0.1,
)

# ── 3A: Counsellor Conversion ────────────────────────────────────────────────
total_by = admission.groupby('attended_by').size()
admit_by = admission[admission['status']=='Admitted'].groupby('attended_by').size()
conv_df  = pd.DataFrame({
    'Counsellor': total_by.index,
    'Total':      total_by.values,
    'Admitted':   admit_by.reindex(total_by.index, fill_value=0).values,
})
conv_df['Conv%'] = (conv_df['Admitted'] / conv_df['Total'] * 100).round(1)
conv_df = conv_df.sort_values('Conv%', ascending=True)

fig_row3.add_trace(go.Bar(
    y=conv_df['Counsellor'], x=conv_df['Total'],
    name='Total Enquiries', orientation='h',
    marker_color='#C7D2FE', opacity=0.6, showlegend=True,
), row=1, col=1)
fig_row3.add_trace(go.Bar(
    y=conv_df['Counsellor'], x=conv_df['Admitted'],
    name='Admitted', orientation='h',
    marker_color='#6366F1', opacity=0.9,
    text=conv_df['Conv%'].apply(lambda v: f'{v:.0f}%'),
    textposition='outside', showlegend=True,
), row=1, col=1)
fig_row3.update_layout(barmode='overlay')

# ── 3B: Attendance Risk Bands ────────────────────────────────────────────────
risk_bins   = [0, 60, 75, 85, 101]
risk_labels = ['Critical\n<60%', 'Warning\n60-75%', 'Regular\n75-85%', 'Excellent\n>85%']
risk_colors = ['#EF4444', '#F59E0B', '#06B6D4', '#10B981']
att_risk = pd.cut(
    attendance['Attendance %'], bins=risk_bins,
    labels=risk_labels, right=False
).value_counts().reindex(risk_labels).fillna(0).reset_index()
att_risk.columns = ['Risk Band', 'Students']

fig_row3.add_trace(go.Bar(
    x=att_risk['Risk Band'], y=att_risk['Students'],
    marker_color=risk_colors,
    text=att_risk['Students'].astype(int),
    textposition='outside',
    showlegend=False,
    hovertemplate='%{x}<br>Students: %{y}<extra></extra>',
), row=1, col=2)

# ── 3C: Staff Workload ───────────────────────────────────────────────────────
staff_ov = staff[staff['batches_handling'] > 0].copy()
staff_ov['capacity_pct'] = (staff_ov['batches_handling'] / 3 * 100).round(0)
staff_ov = staff_ov.sort_values('capacity_pct', ascending=True)

fig_row3.add_trace(go.Bar(
    y=staff_ov['name'], x=[100]*len(staff_ov),
    orientation='h', marker_color='#EEF2FF',
    showlegend=False, hoverinfo='skip',
), row=1, col=3)
fig_row3.add_trace(go.Bar(
    y=staff_ov['name'], x=staff_ov['capacity_pct'],
    orientation='h',
    marker=dict(
        color=staff_ov['capacity_pct'].tolist(),
        colorscale=[[0,'#C7D2FE'],[0.5,'#6366F1'],[1,'#312E81']],
        showscale=False,
    ),
    text=staff_ov['batches_handling'].apply(lambda v: f'{int(v)} batches'),
    textposition='inside', insidetextanchor='middle',
    showlegend=False,
    hovertemplate='%{y}<br>Load: %{x:.0f}%<extra></extra>',
), row=1, col=3)

fig_row3.update_layout(
    barmode='overlay',
    title=dict(text='Row 3 — Conversion · Risk · Workload', font=dict(size=14, color='#1E1B4B'), x=0),
    legend=dict(orientation='h', y=1.1, x=0),
)
plotly_defaults(fig_row3, 340)
fig_row3.update_xaxes(range=[0, 115], row=1, col=3)
fig_row3.show()
```

## 5 · Row 4 — Location · Classroom · Package × Attendance
- **Student Intake by Location** — admitted vs pending stacked with conversion %
- **Classroom Slot Utilisation** — occupied vs free per time slot
- **Package vs Attendance Scatter** — placed students by attendance % and salary

```python

# ═══════════════════════════════════════════════════════════════════════════════
# ROW 4 — Location Intake · Classroom Slots · Package vs Attendance
# ═══════════════════════════════════════════════════════════════════════════════

fig_row4 = make_subplots(
    rows=1, cols=3,
    subplot_titles=[
        '📍 Student Intake by Location',
        '🏫 Classroom Slot Utilisation',
        '💰 Package vs Attendance %',
    ],
    column_widths=[0.38, 0.28, 0.34],
    horizontal_spacing=0.1,
)

# ── 4A: Location stacked bar ─────────────────────────────────────────────────
loc_adm = admission.groupby('location').agg(
    Total=('admission_id','count'),
    Admitted=('status', lambda x: (x=='Admitted').sum())
).reset_index()
loc_adm['Pending'] = loc_adm['Total'] - loc_adm['Admitted']
loc_adm['Conv%']   = (loc_adm['Admitted']/loc_adm['Total']*100).round(1)
loc_adm = loc_adm.sort_values('Total', ascending=True)

fig_row4.add_trace(go.Bar(
    y=loc_adm['location'], x=loc_adm['Admitted'],
    name='Admitted', orientation='h',
    marker_color='#10B981',
    text=loc_adm['Conv%'].apply(lambda v: f'{v:.0f}%'),
    textposition='outside',
), row=1, col=1)
fig_row4.add_trace(go.Bar(
    y=loc_adm['location'], x=loc_adm['Pending'],
    name='Pending', orientation='h',
    marker_color='#E5E7EB',
), row=1, col=1)

# ── 4B: Classroom slot utilisation ──────────────────────────────────────────
slot_cols_ov = ['10:00-12:00','12:00-02:00','02:30-04:30','04:30-05:30']
slot_labels  = ['10–12','12–2','2:30–4:30','4:30–5:30']
slot_data = []
for s, lbl in zip(slot_cols_ov, slot_labels):
    used = classroom[s].astype(str).str.strip().replace('', np.nan).notna().sum()
    slot_data.append({'Slot': lbl, 'Used': int(used), 'Free': len(classroom)-int(used)})
slot_df = pd.DataFrame(slot_data)

fig_row4.add_trace(go.Bar(
    x=slot_df['Slot'], y=slot_df['Used'],
    name='Occupied', marker_color='#6366F1',
    text=slot_df['Used'], textposition='inside',
), row=1, col=2)
fig_row4.add_trace(go.Bar(
    x=slot_df['Slot'], y=slot_df['Free'],
    name='Free', marker_color='#EEF2FF',
    text=slot_df['Free'], textposition='inside',
), row=1, col=2)

# ── 4C: Package vs Attendance scatter ────────────────────────────────────────
placed_df  = placement[placement['Status']=='Placed'].copy()
placed_att = placed_df.merge(
    attendance[['student_id','Attendance %','course']],
    on='student_id', how='left'
).dropna(subset=['Package (Rs.L)','Attendance %'])

if len(placed_att):
    course_list = placed_att['course'].unique().tolist() if 'course' in placed_att.columns else []
    color_map   = {c: PALETTE[i % len(PALETTE)] for i, c in enumerate(course_list)}
    for crs in course_list:
        sub = placed_att[placed_att['course'] == crs]
        fig_row4.add_trace(go.Scatter(
            x=sub['Attendance %'], y=sub['Package (Rs.L)'],
            mode='markers', name=crs,
            marker=dict(size=9, color=color_map[crs], opacity=0.75,
                        line=dict(color='white', width=0.5)),
            hovertemplate='%{text}<br>Att: %{x:.0f}%<br>Pkg: Rs.%{y:.1f}L<extra></extra>',
            text=sub['student_name'] if 'student_name' in sub.columns else sub.index.astype(str),
        ), row=1, col=3)

    avg_p = placed_att['Package (Rs.L)'].mean()
    fig_row4.add_hline(y=avg_p, line_dash='dot', line_color='#EF4444',
                       annotation_text=f'Avg Rs.{avg_p:.1f}L',
                       annotation_font_size=9, row=1, col=3)

fig_row4.update_layout(
    barmode='stack',
    title=dict(text='Row 4 — Location · Classroom · Package×Attendance',
               font=dict(size=14, color='#1E1B4B'), x=0),
    legend=dict(orientation='h', y=1.1, x=0),
)
plotly_defaults(fig_row4, 350)
fig_row4.show()
```

## 8 · Cross-Dataset Radar — Top 4 Courses
Five normalised dimensions across all datasets on one spider chart:
Enrollment · Attendance · Placement · Internship · Staff Support

```python

# ═══════════════════════════════════════════════════════════════════════════════
# ROW 5B — Cross-Dataset Radar: Top 4 Courses vs 5 Dimensions
# ═══════════════════════════════════════════════════════════════════════════════

radar_courses = students['course'].value_counts().head(4).index.tolist()
radar_cats    = ['Enrollment', 'Attendance', 'Placement', 'Internship', 'Staff Support']
colors_r      = ['#6366F1','#06B6D4','#10B981','#F59E0B']

fig_radar = go.Figure()
for i, c in enumerate(radar_courses):
    s_c  = students[students['course'] == c]
    a_c  = attendance[attendance['course'] == c]
    p_c  = placement[placement['Course'] == c]
    pl_c = p_c[p_c['Status'] == 'Placed']
    st_c = staff[staff['course'] == c]

    vals = [
        min(100, len(s_c) / 0.73),                               # enroll → norm to 73 max = 100
        a_c['Attendance %'].mean() if len(a_c) else 0,            # attendance %
        len(pl_c) / max(len(p_c), 1) * 100,                       # placement %
        (s_c['internship'] != 'No').sum() / max(len(s_c), 1) * 100, # internship %
        min(100, len(st_c) * 33),                                  # staff support
    ]
    vals = [round(v, 1) for v in vals]
    short_name = c.replace(' Development','').replace(' with Python','')

    fig_radar.add_trace(go.Scatterpolar(
        r=vals + [vals[0]],
        theta=radar_cats + [radar_cats[0]],
        fill='toself',
        name=short_name,
        line_color=colors_r[i],
        opacity=0.6,
        hovertemplate='%{theta}: %{r:.1f}%<extra></extra>',
    ))

fig_radar.update_layout(
    polar=dict(
        radialaxis=dict(visible=True, range=[0, 100], gridcolor='#E0E7FF',
                        tickfont=dict(size=9)),
        angularaxis=dict(gridcolor='#E0E7FF'),
    ),
    title=dict(text='🕸️ Cross-Dataset Radar — Top 4 Courses (5 Dimensions)',
               font=dict(size=14, color='#1E1B4B'), x=0.5),
    legend=dict(orientation='h', y=-0.15, x=0.5, xanchor='center'),
    showlegend=True,
    paper_bgcolor='white',
    margin=dict(t=60, b=80, l=40, r=40),
)
plotly_defaults(fig_radar, 480)
fig_radar.show()
```
