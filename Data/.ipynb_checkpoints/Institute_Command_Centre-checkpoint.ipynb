## 0. Setup & Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec                      
from matplotlib.colors import LinearSegmentedColormap
import warnings
warnings.filterwarnings('ignore')

# ── Style ──────────────────────────────────────────────────────────────────
plt.rcParams.update({
    'figure.facecolor': '#f0f2ff',
    'axes.facecolor':   '#ffffff',
    'axes.edgecolor':   '#e5e7eb',
    'axes.labelcolor':  '#1e1b4b',
    'axes.titlecolor':  '#1e1b4b',
    'axes.titlesize':   12,
    'axes.labelsize':   10,
    'xtick.color':      '#6b7280',
    'ytick.color':      '#6b7280',
    'grid.color':       '#f3f4f6',
    'grid.linestyle':   '--',
    'font.family':      'DejaVu Sans',
    'legend.framealpha': 0.9,
    'legend.fontsize':   9,
})

COLORS = {
    'primary':   '#4f46e5',
    'teal':      '#06b6d4',
    'green':     '#10b981',
    'amber':     '#f59e0b',
    'red':       '#ef4444',
    'purple':    '#8b5cf6',
    'pink':      '#ec4899',
    'orange':    '#f97316',
    'cyan':      '#14b8a6',
}
PALETTE = list(COLORS.values())

print("✅ Libraries imported and style configured.")
```

## 1. Load Datasets

```python
admission   = pd.read_csv('admission.csv')
attendance  = pd.read_csv('attendance_current_month.csv')
classroom   = pd.read_csv('classroom.csv')
courses     = pd.read_csv('course_dataset.csv')
placement   = pd.read_csv('placement.csv')
staff       = pd.read_csv('staff_dataset.csv')
schedule    = pd.read_csv('staff_schedule_with_weekly.csv')
students    = pd.read_csv('students.csv')

print(f"📋 Admissions  : {len(admission):>4} rows")
print(f"📊 Attendance  : {len(attendance):>4} rows")
print(f"🏫 Classrooms  : {len(classroom):>4} rows")
print(f"📚 Courses     : {len(courses):>4} rows")
print(f"🎯 Placement   : {len(placement):>4} rows")
print(f"👩‍🏫 Staff       : {len(staff):>4} rows")
print(f"📅 Schedule    : {len(schedule):>4} rows")
print(f"🎓 Students    : {len(students):>4} rows")
```

## 2. 📌 KPI Summary Dashboard

```python
# ── Compute KPIs ────────────────────────────────────────────────────────────
total_students   = len(students)
admitted         = (admission['status'] == 'Admitted').sum()
total_enquiries  = len(admission)
adm_rate         = round(admitted / total_enquiries * 100, 1)

placed_count     = (placement['status'] == 'Placed').sum()
placed_pct       = round(placed_count / total_students * 100, 1)

attendance['Attendance %'] = pd.to_numeric(attendance['Attendance %'], errors='coerce')
avg_att          = round(attendance['Attendance %'].mean(), 1)
low_att          = (attendance['Attendance %'] < 75).sum()

pkgs = pd.to_numeric(placement['package(Rs.L)'], errors='coerce').dropna()
avg_pkg          = round(pkgs.mean(), 1)

total_batches    = courses['total_batches'].astype(int).sum()

kpis = [
    ('🎓 Students',       total_students,  '#4f46e5'),
    ('📋 Adm. Rate',      f'{adm_rate}%',  '#06b6d4'),
    ('🎯 Placed',         f'{placed_pct}%','#10b981'),
    ('📊 Avg Att.',       f'{avg_att}%',   '#f59e0b'),
    ('⚠️ <75% Att.',      low_att,         '#ef4444'),
    ('💰 Avg Pkg (L)',    f'₹{avg_pkg}',   '#8b5cf6'),
    ('🏫 Room Fill',      '100%',          '#10b981'),
    ('📚 Batches',        total_batches,   '#f59e0b'),
]

fig, axes = plt.subplots(1, 8, figsize=(22, 2.8))
fig.patch.set_facecolor('#3a2fbf')
for ax, (label, value, color) in zip(axes, kpis):
    ax.set_facecolor('#ffffff')
    for spine in ax.spines.values():
        spine.set_edgecolor(color); spine.set_linewidth(2.5)
    ax.set_xticks([]); ax.set_yticks([])
    ax.text(0.5, 0.62, str(value), ha='center', va='center',
            fontsize=20, fontweight='bold', color=color, transform=ax.transAxes)
    ax.text(0.5, 0.22, label, ha='center', va='center',
            fontsize=8, color='#6b7280', fontweight='600', transform=ax.transAxes)
    ax.axhline(y=0.92, color=color, linewidth=4, xmin=0, xmax=1)

plt.suptitle('Institute Command Centre — April 2026', fontsize=13, fontweight='bold',
             color='white', y=1.08)
plt.tight_layout(pad=0.4)
plt.savefig('kpi_strip.png', dpi=150, bbox_inches='tight', facecolor='#3a2fbf')
plt.show()
print(f"Total Enquiries: {total_enquiries} | Admitted: {admitted} | Placed: {placed_count}")
```

## 3. 📋 Admission Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 3a. Counsellor Conversion
counsellor = admission.groupby('attended_by')['status'].apply(
    lambda x: (x == 'Admitted').sum() / len(x) * 100).sort_values(ascending=True)
bars = axes[0].barh(counsellor.index, counsellor.values,
                     color=[COLORS['primary'], COLORS['teal'], COLORS['green'], COLORS['amber']])
axes[0].set_xlabel('Conversion Rate (%)')
axes[0].set_title('🎯 Counsellor Conversion Rate', fontweight='bold')
axes[0].set_xlim(0, 60)
for bar, val in zip(bars, counsellor.values):
    axes[0].text(val + 0.5, bar.get_y() + bar.get_height()/2,
                 f'{val:.1f}%', va='center', fontweight='bold', fontsize=10)
axes[0].grid(axis='x', alpha=0.4)

# 3b. Intake by Location
loc = admission.groupby('location').size().sort_values(ascending=False)
colors_loc = [COLORS['primary'], COLORS['teal'], COLORS['green'],
              COLORS['amber'], COLORS['purple'], COLORS['pink'], COLORS['orange']]
axes[1].bar(loc.index, loc.values, color=colors_loc, edgecolor='white', linewidth=0.8)
axes[1].set_xlabel('Location')
axes[1].set_title('📍 Enquiries by Location', fontweight='bold')
axes[1].set_ylabel('Count')
for i, (loc_name, val) in enumerate(loc.items()):
    axes[1].text(i, val + 0.8, str(val), ha='center', fontweight='bold', fontsize=9)
axes[1].grid(axis='y', alpha=0.4)

# 3c. Course Enquiries
course_enq = admission.groupby('course').size().sort_values()
axes[2].barh(course_enq.index, course_enq.values, color=COLORS['purple'], alpha=0.85)
axes[2].set_xlabel('Enquiries')
axes[2].set_title('📚 Enquiries by Course', fontweight='bold')
for i, val in enumerate(course_enq.values):
    axes[2].text(val + 0.5, i, str(val), va='center', fontsize=9, fontweight='bold')
axes[2].grid(axis='x', alpha=0.4)

plt.tight_layout()
plt.savefig('admission_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. 📊 Attendance Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 4a. Risk Band Donut
status_counts = attendance['Status'].value_counts()
risk_colors   = {'Excellent': '#10b981', 'Regular': '#3b82f6',
                 'Warning': '#f59e0b', 'Critical': '#ef4444'}
colors_ordered = [risk_colors[s] for s in status_counts.index]
wedges, texts, autotexts = axes[0].pie(
    status_counts.values, labels=status_counts.index,
    colors=colors_ordered, autopct='%1.0f%%',
    pctdistance=0.75, startangle=90,
    wedgeprops={'width': 0.55, 'edgecolor': 'white', 'linewidth': 2})
for at in autotexts: at.set_fontweight('bold'); at.set_fontsize(9)
axes[0].set_title('⚠️ Attendance Risk Bands', fontweight='bold')
axes[0].text(0, 0, f'{avg_att}%\nAvg', ha='center', va='center',
             fontsize=12, fontweight='bold', color='#1e1b4b')

# 4b. Avg Attendance by Course
att_course = attendance.groupby('course')['Attendance %'].mean().sort_values()
bar_colors = ['#ef4444' if v < 75 else '#f59e0b' if v < 85 else '#10b981'
              for v in att_course.values]
axes[1].barh(att_course.index, att_course.values, color=bar_colors)
axes[1].axvline(75, color='#ef4444', linestyle='--', linewidth=1.5, label='75% threshold')
axes[1].axvline(85, color='#f59e0b', linestyle='--', linewidth=1.5, label='85% threshold')
axes[1].set_xlabel('Average Attendance (%)')
axes[1].set_title('📚 Avg Attendance by Course', fontweight='bold')
axes[1].legend(fontsize=8)
for i, val in enumerate(att_course.values):
    axes[1].text(val + 0.3, i, f'{val:.1f}%', va='center', fontsize=8, fontweight='bold')
axes[1].set_xlim(0, 110)
axes[1].grid(axis='x', alpha=0.4)

# 4c. Attendance Distribution Histogram
axes[2].hist(attendance['Attendance %'].dropna(), bins=20,
             color=COLORS['primary'], alpha=0.85, edgecolor='white')
axes[2].axvline(75, color='#ef4444', linestyle='--', linewidth=2, label='75% risk line')
axes[2].axvline(avg_att, color='#10b981', linestyle='-', linewidth=2,
                label=f'Avg: {avg_att}%')
axes[2].set_xlabel('Attendance %')
axes[2].set_ylabel('Number of Students')
axes[2].set_title('📈 Attendance Distribution', fontweight='bold')
axes[2].legend()
axes[2].grid(alpha=0.4)

plt.tight_layout()
plt.savefig('attendance_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. 🏅 Course Health Scorecard

```python
# Build merged scorecard
att_course_avg = attendance.groupby('course')['Attendance %'].mean()

placement['package(Rs.L)'] = pd.to_numeric(placement['package(Rs.L)'], errors='coerce')
placed_df = placement[placement['status'] == 'Placed']
pkg_avg  = placed_df.groupby('course')['package(Rs.L)'].mean()

place_rate = placement.groupby('course').apply(
    lambda x: (x['status'] == 'Placed').sum() / len(x) * 100)

student_cnt = students.groupby('course').size()

scorecard = pd.DataFrame({
    'Students':   student_cnt,
    'Avg Att %':  att_course_avg.round(1),
    'Placed %':   place_rate.round(1),
    'Avg Pkg (L)': pkg_avg.round(1),
}).dropna()

scorecard['Health Score'] = (
    scorecard['Avg Att %'] * 0.40 +
    scorecard['Placed %']  * 0.40 +
    (scorecard['Avg Pkg (L)'] / 20 * 100).clip(0, 100) * 0.20
).round(0).astype(int)

def health_label(score):
    if score >= 70: return 'Excellent'
    if score >= 50: return 'Good'
    if score >= 35: return 'Fair'
    return 'Needs Attention'

scorecard['Health'] = scorecard['Health Score'].apply(health_label)
scorecard = scorecard.sort_values('Health Score', ascending=False)

print(scorecard.to_string())
print()

# Visualise
fig, ax = plt.subplots(figsize=(14, 5))
ax.axis('off')
colors_map = {'Excellent': '#d1fae5', 'Good': '#dbeafe',
              'Fair': '#fef3c7', 'Needs Attention': '#fee2e2'}
cell_colors = [['#f8f9ff'] * 7 for _ in range(len(scorecard))]
for i, (_, row) in enumerate(scorecard.iterrows()):
    cell_colors[i][6] = colors_map.get(row['Health'], '#f8f9ff')

table = ax.table(
    cellText=scorecard[['Students','Avg Att %','Placed %','Avg Pkg (L)','Health Score','Health']].values,
    rowLabels=[c[:25] for c in scorecard.index],
    colLabels=['Students','Avg Att%','Placed%','Avg Pkg(L)','Score','Health'],
    cellLoc='center', loc='center',
    cellColours=cell_colors,
)
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1, 1.6)
ax.set_title('🏅 Course Health Scorecard', fontsize=13, fontweight='bold', pad=10)
plt.tight_layout()
plt.savefig('course_health_scorecard.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 9. 💰 Package vs Attendance (Bubble Chart)

```python
fig, ax = plt.subplots(figsize=(12, 7))

bubble_data = scorecard.copy()
bubble_data['Avg Pkg (L)'] = bubble_data['Avg Pkg (L)'].fillna(0)

for i, (course, row) in enumerate(bubble_data.iterrows()):
    size = row['Students'] * 12
    color = PALETTE[i % len(PALETTE)]
    ax.scatter(row['Avg Att %'], row['Avg Pkg (L)'], s=size,
               color=color, alpha=0.75, edgecolor='white', linewidth=2, zorder=3)
    short = course[:18] + ('…' if len(course) > 18 else '')
    ax.annotate(short, (row['Avg Att %'], row['Avg Pkg (L)']),
                textcoords='offset points', xytext=(6, 4), fontsize=8, fontweight='500')

ax.axvline(75, color='#ef4444', linestyle='--', alpha=0.6, label='75% attendance line')
ax.axhline(bubble_data['Avg Pkg (L)'].mean(), color='#4f46e5',
           linestyle='--', alpha=0.6, label=f'Avg pkg ₹{bubble_data["Avg Pkg (L)"].mean():.1f}L')
ax.set_xlabel('Average Attendance (%)', fontsize=11)
ax.set_ylabel('Average Package (₹ Lakhs)', fontsize=11)
ax.set_title('💰 Package vs Attendance — Bubble size ∝ Students', fontsize=13, fontweight='bold')
ax.legend(); ax.grid(alpha=0.4)

plt.tight_layout()
plt.savefig('pkg_vs_attendance.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 10. 🌐 Cross-Dataset Radar — Top 4 Courses

```python
import numpy as np

top4 = scorecard.sort_values('Students', ascending=False).head(4)
categories = ['Attendance %', 'Placed %', 'Pkg (scaled)', 'Students (scaled)', 'Health Score']
N = len(categories)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(9, 9), subplot_kw={'projection': 'polar'})
ax.set_facecolor('#f8f9ff')
fig.patch.set_facecolor('#f0f2ff')

radar_colors = ['#4f46e5', '#10b981', '#f59e0b', '#ef4444']

for i, (course, row) in enumerate(top4.iterrows()):
    values = [
        row['Avg Att %'],
        row['Placed %'],
        min(row['Avg Pkg (L)'] / 20 * 100, 100),
        min(row['Students'] / top4['Students'].max() * 100, 100),
        row['Health Score'],
    ]
    values += values[:1]
    short = course[:20] + ('…' if len(course) > 20 else '')
    ax.plot(angles, values, 'o-', linewidth=2, color=radar_colors[i], label=short)
    ax.fill(angles, values, alpha=0.12, color=radar_colors[i])

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, size=10, fontweight='bold')
ax.set_ylim(0, 100)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(['20', '40', '60', '80', '100'], size=8, color='#9ca3af')
ax.grid(color='#e5e7eb', linewidth=0.8)
ax.set_title('🌐 Cross-Dataset Radar — Top 4 Courses', size=14, fontweight='bold', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.35, 1.15))

plt.tight_layout()
plt.savefig('radar_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```
